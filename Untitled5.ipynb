```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
bankdata = pd.read_csv(r"SVMdataset.csv")
```

```python

X = bankdata.iloc[:, :-1].values
Y = bankdata.iloc[:, -1].values


from sklearn.model_selection import train_test_split


x, x_test, y, y_test = train_test_split(X,Y,test_size=0.2,train_size=0.8)
x_train, x_cv, y_train, y_cv = train_test_split(x,y,test_size = 0.25,train_size =0.75)
```

```python
X1=bankdata['x1']
X2=bankdata['x2']
X_training=np.array(list(zip(X1,X2)))
type(X_training)
```

```python
y_training=bankdata['y']
type(y_training)
```

```python
target_names=['-ve','+ve']
target_names
```

```python
idxPlus=y_training[y_training<=0].index
idxMin=y_training[y_training>0].index
plt.scatter(X_training[idxPlus,0],X_training[idxPlus,1],c='r',s=30)
plt.scatter(X_training[idxMin,0],X_training[idxMin,1],c='b',s=30,marker='x')
plt.legend(target_names,loc=2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('SVM Dataset')
plt.savefig('chart0.png')
```

```python
new_y_cv= pd.Series(y_cv)

idxPlus=new_y_cv[new_y_cv<=0].index
idxMin=new_y_cv[new_y_cv>0].index
plt.scatter(x_cv[idxPlus,0],x_cv[idxPlus,1],c='r',s=30)
plt.scatter(x_cv[idxMin,0],x_cv[idxMin,1],c='b',s=30,marker='x')
plt.legend(target_names,loc=2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('SVM Dataset')
plt.savefig('chart0.png')
```

```python
new_y_train= pd.Series(y_train)

idxPlus=new_y_train[new_y_train<=0].index
idxMin=new_y_train[new_y_train>0].index
plt.scatter(x_train[idxPlus,0],x_train[idxPlus,1],c='r',s=30)
plt.scatter(x_train[idxMin,0],x_train[idxMin,1],c='b',s=30,marker='x')
plt.legend(target_names,loc=2)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('SVM Dataset')
plt.savefig('chart0.png')
```

```python
from sklearn import svm

from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
best_C = None
best_gamma = None
best_score=0
for C in np.arange(10, 15, 20):
    for gamma in np.arange(0.001, 0.101, 0.001):
        model = SVC(kernel='linear', gamma=gamma, C=C)
        model.fit(x_train, y_train)
        score = model.score(x_cv, y_cv)
        if ((score)>(best_score)):
            best_score = score
            best_C = C
            best_gamma = gamma
print('Highest Accuracy Score: ', best_score)  
print(best_C)
print(best_gamma)
```
